## Treating missing values

```python
# Calculate the threshold for >30% missing values
threshold = (30 / 100) * len(df)

# List columns with more than 30% missing values
columns_to_impute = df.columns[df.isnull().sum() > threshold]

# Impute missing values for columns with more than 30% missing values
for column in columns_to_impute:
    if df[column].dtype in ['int64', 'float64']:  # Check if the column is numerical
        df[column].fillna(df[column].mean(), inplace=True)
    else:
        df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
# Calculate the threshold for <30% missing values
threshold = (30 / 100) * len(df)

# Identify columns with less than 30% missing values
columns_to_impute = df.columns[df.isnull().sum() < threshold]

# Impute missing values for columns with less than 30% missing values
for column in columns_to_impute:
    if df[column].dtype in ['int64', 'float64']:  # Check if the column is numerical
        df[column].fillna(df[column].mean(), inplace=True)
    else:
        df[column].fillna(df[column].mode()[0], inplace=True) 
```

## Task 5 -

<font size = '4'> Compare the performance of cars based on factors like kerb weight and gross vehicle weight.<font/>

<font size = '4'> Hint: Calculate and display the mean, median, standard deviation, and coefficient of variation for each manufacturer.<font/>

```python
# Convert 'Kerb_Weight' and 'Gross_Vehicle_Weight' columns to string type
df['Kerb_Weight'] = df['Kerb_Weight'].astype(str)
df['Gross_Vehicle_Weight'] = df['Gross_Vehicle_Weight'].astype(str)

# Remove 'kg' suffix from the Kerb_Weight and Gross_Vehicle_Weight columns
df['Kerb_Weight'] = df['Kerb_Weight'].str.replace('kg', '')
df['Gross_Vehicle_Weight'] = df['Gross_Vehicle_Weight'].str.replace('kg', '')

# Split values separated by space and take the first part
import pandas as pd

# Assuming 'df' contains your DataFrame

# Convert 'Kerb_Weight' and 'Gross_Vehicle_Weight' columns to string type
df['Kerb_Weight'] = df['Kerb_Weight'].astype(str)
df['Gross_Vehicle_Weight'] = df['Gross_Vehicle_Weight'].astype(str)

# Remove 'kg' suffix from the Kerb_Weight and Gross_Vehicle_Weight columns
df['Kerb_Weight'] = df['Kerb_Weight'].str.replace('kg', '')
df['Gross_Vehicle_Weight'] = df['Gross_Vehicle_Weight'].str.replace('kg', '')

# Split values separated by space and take the first part
df['Kerb_Weight'] = df['Kerb_Weight'].str.split().str[0]
df['Gross_Vehicle_Weight'] = df['Gross_Vehicle_Weight'].str.split().str[0]

# Convert the columns to numeric data type
df['Kerb_Weight'] = pd.to_numeric(df['Kerb_Weight'], errors='coerce')
df['Gross_Vehicle_Weight'] = pd.to_numeric(df['Gross_Vehicle_Weight'], errors='coerce')

# Define the columns to analyze
columns_to_analyze = ['Kerb_Weight', 'Gross_Vehicle_Weight']

# Define a function to calculate the coefficient of variation
def coefficient_of_variation(data):
    mean = data.mean()
    std_dev = data.std()
    return (std_dev / mean) * 100

# Define a function to display the statistics
def display_statistics(df):
    statistics = pd.DataFrame()
    statistics['Mean'] = df.mean()  # Calculate mean
    statistics['Median'] = df.median()  # Calculate median
    statistics['Standard Deviation'] = df.std()  # Calculate standard deviation
    statistics['Coefficient of Variation'] = coefficient_of_variation(df)  # Calculate coefficient of variation
    return statistics

# Group by 'Make' and calculate statistics for each manufacturer
grouped_statistics = df.groupby('Make')[columns_to_analyze].apply(display_statistics)

# Display the grouped statistics
print(grouped_statistics.head().round(2))


```

## *Interpretation:*

-  <font size = '4'> This analysis helps to understand the weight distribution of cars among different manufacturers, for which calculated the average(mean), meadian and variability(standard deviation & coefficient of variation) of **Kerb_Weight and Gross_Vehicle_Weight**.<font/>
    
    
-  <font size = '4'> The car manufacturer - **Audi** had the highest weight compared to all other manufacturers related to vehicle weight.<font/>

## 1. Safety Feature Assessment

## Task 1 - 

<font size = '4'> Examine the presence of safety features such as ABS, airbags, and hill assist, to assess the safety standards of various car models.<font/>

```python
# Define the safety features columns
safety_features_columns = ['ABS_(Anti-lock_Braking_System)', 'Airbags', 'Hill_Assist']

# Define a function to convert 'Yes' and 'No' to binary values
def convert_to_binary(value):
    if value == 'Yes':
        return 1
    elif value == 'No':
        return 0
    else:
        return None

# Convert safety features columns to binary
for col in safety_features_columns:
    df[col] = df[col].apply(convert_to_binary)

# Define a function to display the presence of safety features statistics
def display_safety_features_statistics(df):
    statistics = pd.DataFrame()
    statistics['Presence'] = df.sum()  # Count the presence of each safety feature
    statistics['Percentage Presence'] = (df.sum() / df.shape[0]) * 100  # Calculate the percentage presence of each safety feature
    return statistics

# Group by 'Make' and calculate safety features statistics for each manufacturer
grouped_safety_features_statistics = df.groupby('Make')[safety_features_columns].apply(display_safety_features_statistics)

# Display the grouped safety features statistics
print(grouped_safety_features_statistics)

```

## *Interpretation:*

-  <font size = '4'> This analysis defines the safety features based on **ABS_(Anti-lock_Braking_System), Airbags, and Hill_Assist**, in order to calculate the statistics related to the presence of safety features,converted values to binary i.e.,'Yes' and 'No' values to binary values (1 for 'Yes' and 0 for 'No').<font/>

-  <font size = '4'> Calculated the statistics for safety features which include count and presence for each manufacturer.<font/>

-  <font size = '4'> The car manufacturer - **Maruti Suzuki** has the presence highest count of ABS_(Anti-lock_Braking_System), Airbags, and Hill_Assist compared to other car manufacturers.<font/>

<font size = '4'> Investigate the presence of high-speed alerts, seat belt reminders, and door ajar warnings to assess the integration of safety and convenience features.<font/>

```python
# Group the data by car model and determine if each safety feature is present for each model
feature_presence_by_model = df.groupby('Model')[['High_Speed_Alert_System', 'Fasten_Seat_Belt_Warning', 'Door_Ajar_Warning']].apply(lambda x: (x == 'Yes').any()).reset_index()

# Calculate the total count of car models that have each feature
feature_presence_statistics = feature_presence_by_model.drop('Model', axis=1).sum()
total_car_models = len(feature_presence_by_model)

# Calculate the percentage of car models that have each feature
feature_presence_percentage = (feature_presence_statistics / total_car_models) * 100

# Create a bar plot to visualize the percentage of car models with each feature
plt.figure(figsize=(12, 8))
barplot = sns.barplot(x=feature_presence_percentage.index, y=feature_presence_percentage.values, palette='viridis')
plt.title('Presence of Safety and Convenience Features Across Car Models')
plt.xlabel('Car Model')
plt.ylabel('Percentage of Car Models with Feature')
plt.xticks(rotation=45)

# Display percentage values at the top of each bar along with car model names
for index, (model, value) in enumerate(zip(feature_presence_by_model['Model'], feature_presence_percentage)):
    plt.text(index, value, f'{model}\n{value:.2f}%', ha='center', va='bottom')

plt.tight_layout()
plt.show()

```
