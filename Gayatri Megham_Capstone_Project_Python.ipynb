## Changing Special Character from Games dataset

```python
def replace_special_characters_with_alphabets(games):
    columns_to_process = ['game_id', 'season', 'home_club_goals', 'away_club_goals',
                           'home_club_position', 'away_club_position', 'attendance',
                           'competition_id','round', 'home_club_manager_name', 'away_club_manager_name', 
                           'stadium', 'referee', 'home_club_name', 'away_club_name', 
                           'aggregate', 'competition_type']

    # Define a mapping of special characters to their replacement alphabets
    special_char_mapping = {
        '!': 'A', '@': 'B', '#': 'C', '$': 'D', '%': 'E', '^': 'F', '&': 'G', '*': 'H', '(': 'I', ')': 'J',
        '1/4': '', 'Ã': 'A', '¶': '', '¦': '', '¸': ''
    }

    # Function to replace a single character based on the mapping
    def replace_char(char):
        return special_char_mapping.get(char, char)  # If char not in mapping, keep it unchanged

    # Iterate through each specified column and apply the replacement
    for column in columns_to_process:
        games[column] = games[column].apply(lambda x: ''.join(replace_char(char) for char in str(x)))

    return games


# Replace special characters in the specified columns
games = replace_special_characters_with_alphabets(games)

# Now 'games' contains the modified data with special characters replaced

games.head()
```

# Univariate Analysis for each dataset

## 1. Univariate analysis for Games-Dataset

## Categorical Columns

```python
# Assuming your DataFrame is named game_1
categorical_columns = ['competition_id', 'round', 'home_club_manager_name',
                        'away_club_manager_name', 'stadium', 'referee',
                        'home_club_name', 'away_club_name',  'competition_type']

# Set the threshold for considering top N categories
top_categories_threshold = 10

# Display bar charts for each categorical column based on unique values
for column in categorical_columns:
    unique_values = games[column].nunique()

    plt.figure(figsize=(10, 5))
    
    if unique_values <= top_categories_threshold:
        # If the number of unique values is small, use a regular bar chart with count annotations
        value_counts = games[column].value_counts()
        value_counts.plot(kind='bar', color='orange')
        
        # Display count values on top of the bars
        for i, value in enumerate(value_counts):
            plt.text(i, value + 0.1, str(value), ha='center', va='bottom')
        
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Count')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()

    else:
        # If the number of unique values is large, take the top N categories and use a horizontal bar chart
        top_categories = games[column].value_counts().nlargest(top_categories_threshold).index
        games[column] = games[column].apply(lambda x: x if x in top_categories else 'Others')
        
        value_counts = games[column].value_counts()
        value_counts.plot(kind='barh', color='red')
        
        # Display count values on top of the bars
        for i, value in enumerate(value_counts):
            plt.text(value + 0.1, i, str(value), ha='left', va='center')
        
        plt.title(f'Distribution of {column} (Top {top_categories_threshold} categories)')
        plt.xlabel('Count')
        plt.ylabel(column)
        plt.tight_layout()

    plt.show()
```

## Numerical Columns

```python
# Assuming your DataFrame is named game_1
numeric_columns = ['game_id', 'season', 'home_club_goals', 'away_club_goals',
                   'home_club_position', 'away_club_position', 'attendance']

# Convert columns to numeric type if possible
games[numeric_columns] = games[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Set the number of subplots and their layout
num_subplots = len(numeric_columns)
num_rows = (num_subplots + 1) // 2  # Calculate the number of rows

# Create subplots with a consistent color for histograms
fig, axes = plt.subplots(num_rows, 2, figsize=(15, 5 * num_rows))

# Flatten the 2D array of subplots for easier indexing
axes = axes.flatten()

# Plot histograms for each numeric column
for i, column in enumerate(numeric_columns):
    ax = sns.histplot(data=games, x=column, bins=20, color='black', ax=axes[i], edgecolor='black')
    
    # Display percentage values on top of the bars without overlapping
    total_count = games[column].dropna().count()
    for p in ax.patches:
        percentage = p.get_height() / total_count * 100
        if percentage > 0:
            ax.annotate(f'{percentage:.2f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                         ha='center', va='bottom', fontsize=8, color='red', xytext=(0, 5),
                         textcoords='offset points')

    ax.set_title(f'Histogram for {column}')
    ax.set_xlabel(column)
    ax.set_ylabel('Frequency')

# Remove empty subplot if the number of subplots is odd
if num_subplots % 2 != 0:
    fig.delaxes(axes[-1])

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```
