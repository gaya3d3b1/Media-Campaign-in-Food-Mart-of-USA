# **Task 2 - Booking Pattern Analysis**

## 2.1 - What are the peak booking periods for the hotel?

```python
# Convert arrival_date_month to a categorical type with correct order
df['arrival_date_month'] = pd.Categorical(df['arrival_date_month'], categories=[
    'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'
], ordered=True)

# Group by year and month to count the number of bookings
monthly_bookings = df.groupby(['arrival_date_year', 'arrival_date_month'])['hotel'].count().reset_index()

# Calculate total bookings per month across all years
monthly_total = monthly_bookings.groupby('arrival_date_month')['hotel'].sum()

# Plot the monthly booking trend
plt.figure(figsize=(12, 6))
plt.plot(monthly_total, marker='o', color='blue', linestyle='-')
plt.title('Monthly Booking Trend')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()



```

## **Interpretation:** 

<font size = '4'> This analysis represents the monthly booking trends.<font/>


-  <font size = '4'> The month of **July** has the highest booking trend followed by **June & August**.<font/>

-  <font size = '4'> The month of **November** has the lowest booking trend followed by **February & December**.<font/>

## 2.3 - How do booking patterns vary by month, week, and day of the week?

```python
# Booking Patterns by Month (Line Plot)
monthly_bookings = df.groupby('arrival_date_month').size().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
plt.plot(monthly_bookings.index, monthly_bookings.values, marker='o', color='#483d8b')
plt.title('Booking Patterns by Month')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Group by week number and count the number of bookings
weekly_bookings = df.groupby('arrival_date_week_number').size()

# Plotting booking patterns by week number (line plot)
plt.figure(figsize=(12, 6))
weekly_bookings.plot(color='#f29e4c')
plt.title('Booking Patterns by Week Number')
plt.xlabel('Week Number')
plt.ylabel('Number of Bookings')
plt.grid(True)
plt.tight_layout()
plt.show()

# Convert columns to appropriate data types if needed
df['arrival_date_year'] = df['arrival_date_year'].astype(str)
df['arrival_date_month'] = df['arrival_date_month'].astype(str)
df['arrival_date_day_of_month'] = df['arrival_date_day_of_month'].astype(str)

# Concatenate columns to create a date string
date_str = df['arrival_date_year'] + '-' + df['arrival_date_month'] + '-' + df['arrival_date_day_of_month']

# Convert the concatenated string to datetime
df['arrival_date'] = pd.to_datetime(date_str)

# Extract the day of the week
df['arrival_date_day_of_week'] = df['arrival_date'].dt.dayofweek


# Group by day of the week and count the number of bookings
day_of_week_bookings = df['arrival_date_day_of_week'].value_counts().sort_index()

# Plotting booking patterns by day of the week (bar plot)
plt.figure(figsize=(10, 6))
day_of_week_bookings.plot(kind='bar', color='#be4f62')
plt.title('Booking Patterns by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 3.2 - Are there specific trends or patterns in cancellations over time?

```python
# Convert arrival_date_month to a categorical type with correct order
df['arrival_date_month'] = pd.Categorical(df['arrival_date_month'], 
                                           categories=['January', 'February', 'March', 'April', 'May', 'June', 
                                                       'July', 'August', 'September', 'October', 'November', 'December'],
                                           ordered=True)

# Group by arrival date month and calculate cancellation rates
cancellation_by_month = df.groupby(['arrival_date_month', 'is_canceled']).size().unstack()
cancellation_by_month['cancellation_rate'] = cancellation_by_month[1] / (cancellation_by_month[0] + cancellation_by_month[1])

# Plot cancellation rates over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=cancellation_by_month['cancellation_rate'], marker='o', color='red')
plt.title('Cancellation Rates Over Time')
plt.xlabel('Month')
plt.ylabel('Cancellation Rate')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 3.3 - Can we identify high-risk bookings that are more likely to be cancelled?

```python
# Explore cancellation rate by different features
plt.figure(figsize=(12, 8))

# Cancellation rate by hotel type
plt.subplot(3, 1, 1)
sns.countplot(x='hotel', hue='is_canceled', data=df)
plt.title('Cancellation Rate by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Count')

# Cancellation rate by market segment
plt.subplot(3, 1, 2)
sns.countplot(x='market_segment', hue='is_canceled', data=df)
plt.title('Cancellation Rate by Market Segment')
plt.xlabel('Market Segment')
plt.ylabel('Count')
plt.xticks(rotation=45)

# Cancellation rate by customer type
plt.subplot(3, 1, 3)
sns.countplot(x='customer_type', hue='is_canceled', data=df)
plt.title('Cancellation Rate by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Count')

plt.tight_layout()
plt.show()

# Calculate cancellation rates by different features
cancellation_by_features = df.groupby(['hotel', 'market_segment', 'customer_type'])['is_canceled'].mean().reset_index()

# Filter high-risk bookings with a cancellation rate above a certain threshold
high_risk_bookings = cancellation_by_features[cancellation_by_features['is_canceled'] > 0.5]

if not high_risk_bookings.empty:
    # Print high-risk bookings
    print("High-risk bookings with cancellation rate > 50%:")
    print(high_risk_bookings)
    
    # Visualize high-risk bookings
    plt.figure(figsize=(12, 6))
    sns.barplot(x='is_canceled', y='hotel', hue='market_segment', data=high_risk_bookings)
    plt.title('High-risk Bookings with Cancellation Rate > 50%')
    plt.xlabel('Cancellation Rate')
    plt.ylabel('Hotel Type')
    plt.legend(title='Market Segment')
    plt.show()
else:
    print("No high-risk bookings found.")

# Analyze cancellation rates for bookings with high lead time
high_lead_time_bookings = df[df['lead_time'] > 180]  # Define high lead time as over 180 days
cancellation_rate_high_lead_time = high_lead_time_bookings['is_canceled'].mean()
print("Cancellation rate for bookings with high lead time:", cancellation_rate_high_lead_time)

# Analyze cancellation rates for bookings made through specific distribution channels
distribution_channel_cancel_rates = df.groupby('distribution_channel')['is_canceled'].mean()
print("Cancellation rates by distribution channel:")
print(distribution_channel_cancel_rates)
```

-  <font size = '4'> The first plot is to display the cancellation rate by hotel type, where **City Hotel** has the highest Non Cancellation and Cancellation rate compared to **Resort Hotel**.<font/>
    
-  <font size = '4'> The second plot is to display the cancellation rate by Market Segment, where **Online TA** has the highest Non Cancellation and Cancellation rate compared to **Resort Hotel**.<font/>

## 1.1 - How does the Average Daily Rate (ADR) vary over time?

```python
# Convert date-related columns to string data type
df['arrival_date_year'] = df['arrival_date_year'].astype(str)
df['arrival_date_month'] = df['arrival_date_month'].astype(str)
df['arrival_date_day_of_month'] = df['arrival_date_day_of_month'].astype(str)

# Concatenate date columns to create a date string
date_str = df['arrival_date_year'] + '-' + df['arrival_date_month'] + '-' + df['arrival_date_day_of_month']

# Convert the concatenated string to datetime
df['arrival_date'] = pd.to_datetime(date_str)

# Group the data by month and calculate the average daily rate (ADR)
monthly_adr = df.groupby(pd.Grouper(key='arrival_date', freq='M'))['adr'].mean()

# Plot the variation of the average daily rate over time
plt.figure(figsize=(12, 6))
monthly_adr.plot(marker='o', color='b', linestyle='-')
plt.title('Average Daily Rate (ADR) Variation Over Time')
plt.xlabel('Date')
plt.ylabel('Average Daily Rate (ADR)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## **Interpretation:**

<font size = '4'> This analysis is to display the average daily rate (ADR) variation over time.<font/>
    
-  <font size = '4'> The month - **July 2017** marked the highest trend/variation over time.<font/>    

-  <font size = '4'> The month - **Jan 2016 & 2017** marked the lowest trend/variation over time.<font/> 

## 1.2 - Can we identify pricing strategies that maximize revenue?

```python
# Calculate total revenue
df['total_revenue'] = df['adr'] * (df['stays_in_weekend_nights'] + df['stays_in_week_nights'])

# Create a scatter plot with trend line
plt.figure(figsize=(10, 6))
sns.regplot(x='adr', y='total_revenue', data=df, color = 'black')
plt.title('Relationship between ADR and Total Revenue')
plt.xlabel('ADR')
plt.ylabel('Total Revenue')
plt.show()
```

## **Interpretation:**


<font size = '4'> This analysis is to display the relationship between ADR and Total Revenue.<font/>
    
    
-  <font size = '4'> The trend line displays the **Weak Positive Correlation** between ADR and Total Revenue generated. <font/>
